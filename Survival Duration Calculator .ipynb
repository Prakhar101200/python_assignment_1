```python
import time #importing time package

def survival_duration_cal(): #defining a function for calculating user's age in various time metrics
    
    print('''
    ****WElCOME TO SURVIVAL DURATION CALCULATOR****''') #Here, first we are giving a welcome message to the user. 
    
    input_age = int(input("please enter your age: ")) #asking user to enter their age 
   # if not isinstance(input_age, float): #giving a if-else condition to alert user to give the valid input format
    #    print("Please enter a valid age")

    print(''' Select your options
    1. months
    2. weeks
    3. days
    4. hours 
    5. minutes
    6. seconds''') #printing the available options of this functions, where user's can choose particular option by choosing among the serial number. 
    
    choose = int(input("Please Select Among the Options to Calculate: ")) #asking user their age, they want to calculate
    if choose == 1:
        age_in_months = input_age * 12 #calculating user's age into months
        print("Your exact age in months is:", age_in_months)
    elif choose == 2: #using  elif function to calculate each function
        week_in_nonleap_year = 52.143 #taking a constant for no. of weeks in a non leap year 
        week_in_leap_year = 52.286 #taking a constant for no. of weeks in a leap year 
        leap_year_count = input_age // 4 #using '//' operator calculating no. of leap year count and only accepting only it's whole number part
        non_leap_year_count = input_age - leap_year_count #now, calulating the no. of the non-leap year by substracting the number of leap-year from the input age
        non_leap_year_weeks = non_leap_year_count * week_in_nonleap_year # calculating no. of weeks in non-leap years
        leap_year_weeks = leap_year_count * week_in_leap_year #calculating no. of weeks in leap years
        age_in_weeks = non_leap_year_weeks + leap_year_weeks #finally calculating the user's age in weeks with taking in account the number of leap years after every fourth year. 
        print("Your age in weeks is:", age_in_weeks) #printing the calculated age in weeks 
    elif choose == 3: #using elif function for calculating the 3rd option
        days_in_nonleap_year = 365 #taking no. of days in a year as constant
        days_in_leap_year = 366 #also, no =. of days in a leap-year as constant 
        leap_year_count = input_age // 4 # using '//' operator to calculate the number of leap year occurred in user age. Considering the leap year occuring after wvery 4th year.
        non_leap_year_count = input_age - leap_year_count #calculating non-leap year of user's input age.
        age_in_leap_years = leap_year_count * days_in_leap_year #calculating age of the user in leap years 
        age_in_non_leap_year = non_leap_year_count * days_in_nonleap_year #calculating age of the user in rest of the non-leap year
        age_in_days = age_in_leap_years + age_in_non_leap_year #adding them to get the total age in days
        print("Your age in days are:", age_in_days) #printing the age in days.
    elif choose == 4: #giving the elif function as if the user wants to calculate their age in hours
        hours_in_non_leap_year = 8760 #taking a constant value of total hours in a non-leap years
        hours_in_leap_year = 8784 #taking a constant value of total hours in a leap years 
        leap_year_count = input_age // 4 #counting the number of leap years in the user's input age by using '//' operator to only get the ans in whole number. 
        non_leap_year_count = input_age - leap_year_count #counting the number of non-leap years 
        days_in_non_leap_year = non_leap_year_count * hours_in_non_leap_year #counting the number of hours in non-leap years 
        days_in_leap_year = leap_year_count * hours_in_leap_year #counting the number of hours in leap years
        age_in_hours = days_in_non_leap_year + days_in_leap_year #adding the total user's age in hours
        print("Your age in days is:", age_in_hours) #printing the amount of hours user has lived.
    elif choose == 5: #giving a elif function, if the user wants to calculate their age in minutes 
        minutes_in_non_leap_year = 525600 #taking a constant value of number of minutes in non-leap years 
        minutes_in_leap_year = 527040 #taking a constant value of number of minutes in leap years
        leap_year_count = input_age // 4 #counting the number of leap years by using '//' operator to get the ans as a whole number. 
        non_leap_year_count = input_age - leap_year_count #counting the number of non-leap years 
        minutes_age_in_non_leap_year = non_leap_year_count * minutes_in_non_leap_year #counting the number of minutes in a non-leap years 
        minutes_age_in_leap_year = leap_year_count * minutes_in_leap_year #counting the number of minutes in a leap year
        age_in_minutes = minutes_age_in_non_leap_year + minutes_age_in_leap_year #now totalling the user's age in minutes.
        print("Your age in minutes are:", age_in_minutes) #printing the final ans 
    elif choose == 6: #giving elif function to let user choose if he/she wants to cal. their age in seconds 
        seconds_in_non_leap_year = 31536000 #taking a constant value of number of seconds in a non-leap year
        seconds_in_leap_year = 31622400 #taking a constant value of number of seconds in a leap year 
        leap_year_count = input_age // 4 #counting the number of leap years by using '//' operator. 
        non_leap_year_count = input_age - leap_year_count #counting the number of non-leap years
        seconds_age_in_non_leap_year = non_leap_year_count * seconds_in_non_leap_year #now, calculating the age of the user in seconds in non-leap years
        seconds_age_in_leap_year = leap_year_count * seconds_in_leap_year #here, we are calculating the age of the user in seconds in leap years 
        age_in_seconds = seconds_age_in_non_leap_year + seconds_age_in_leap_year #here, we are adding both number of seconds and calculating the exact number of age of the user in seconds
        print("Your age in seconds is:", age_in_seconds) #printing the ans
        
    else:  # now if the person doesn't choose any option then using time.exit function in else loop to terminate the code in 10 seconds. 
        time.sleep(10)
        print("No options selected")
        
        exit() #using the exit() function where the code stops working after the given time period.
     
    return


if __name__ == "__main__":
    
    
    survival_duration_cal()
    
    
```
