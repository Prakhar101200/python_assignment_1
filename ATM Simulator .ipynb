```python
import time #importing time package
def atm_simulator(): #defining a function atm simulator
    bank_balance = 100000 #giving the constant for the amount of money present in the account of the user.
    
    print('''*****WELCOME TO THE PG BANK ATM KIOSK*****''') #printing welcome statement
    
    user_pin = int(input("Please enter your pin: ")) #asking the user to inout the pin number 
    if not(1000 <= user_pin <=9999): #checking if the pin number entered is of 4 digits or not
        print("Please re-enter your pin") # giving the print function to tell youu to re-enter your pin, if your entered pin isn't of 4 digits.
        return # using the return function to stop the code if the user enters wrong pin
    
    print('''WELCOME TO THE HOMEPAGE OF PG BANK ATM KIOSK
    CHOOSE ONE OPTION
    1.Check Balance        
    2.Withdrawl
    3.Deposit
    4.Exit''') #giving several performing tasks of this simulator if the user enters the right pin
    
    
    choose = int(input("Choose Any One Option: ")) #asking the user to enter the option number among the above performing tasks lists. 
    if choose == 1: #using if statement to compute the task of each present option.
        print("Available balance in your account:", bank_balance) #giving the print function to show the A/C balance of the user. 
    
    elif choose == 2: #giving the elif function, if the user chooses the seond option, here, it is withdrawl
        input_withdraw = int(input("Enter the withdrawl amount: ")) #asking the user to enter the withdrawing amount. 
        if input_withdraw < bank_balance: #using if statement to give the condition if the input withdrawing amount is less than balance.
            bank_balance -= input_withdraw #if the above statement comes true, updating the A/C balance with diducting the withdrawl amount
            print(f'Your available balance is: {bank_balance}') #printing the updated balance 
        else: #using else statement, if the above condition becomes false 
            print('''Insufficient balance 
            transaction failed''') #if the withdrawing amount exceeds the user gets an message of 'Insufficient balance transaction failed'
    
    elif choose == 3: #using elif function if the user chooses the 3rd option  
        input_deposit = int(input("Please enter depositing amount: ")) #asking the user to enter the depositing amount 
        bank_balance += input_deposit #adding the depositing amount to the available bank balance
        print(f'Your new bank balance is: {bank_balance}') #printing new updated bank balance 
    
    elif choose == 4: #using elif function if the user chooses the 4th option 
        print("Thank you for using PG BANK ATM KIOSK, now you will be exited from the transaction") #giving the message to the user that you exited from the atm simulator
        exit() #using exit function to break the code
        
    else:  #giving else function if the user fails to choose any option 
        time.sleep(10) #using the function that breaks the code after 10 secs if the user doesn't performs any tasks from the simulator
        exit() #exiting the code
    

if __name__ == "__main__":
    
    atm_simulator()
```

Project Review Documentation


The "ATM Simulator" is a Python program that emulates the functionality of an ATM kiosk. This simulator allows users to check their account balance, withdraw money, deposit money, and exit the transaction, providing a virtual banking experience. It ensures that the user input is valid and updates the account balance accordingly.

Logic Build


The algorithm for the "ATM Simulator" is designed to perform basic banking operations based on user input. Here's how the logic works:

The user is welcomed to the ATM kiosk and asked to enter their PIN (Personal Identification Number).

If the entered PIN is not a four-digit number, the user is prompted to re-enter it. If they fail to do so, the program terminates.

Once the PIN is verified, the user is presented with a menu of options: check balance, withdraw, deposit, or exit.

Depending on the user's choice, the program performs the corresponding operation:

If the user selects "Check Balance," the available balance is displayed.
If the user chooses "Withdraw," they are asked to enter the withdrawal amount. If the withdrawal amount is within the account balance, it's deducted, and the updated balance is displayed. Otherwise, the user receives an "Insufficient balance" message.
If "Deposit" is selected, the user is prompted to enter the deposit amount, and it's added to the account balance.
If the user decides to "Exit," the program terminates.
If the user fails to select any option within 10 seconds, the program automatically exits.


Functions and Operators Used

The code employs various Python functions and operators to accomplish its tasks:


print: Displays messages, the menu, and account balance to the user.
input: Gathers user input for the PIN, withdrawal amount, and deposit amount.
Conditional statements (if, elif, else): Determine which action the user wants to perform.
Arithmetic operators (+=, -=): Update the account balance based on user actions.
time.sleep: Pauses the program for 10 seconds if the user doesn't make a selection.
exit(): Terminates the program when the user exits or in the case of inactivity.
